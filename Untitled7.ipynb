```python
import numpy as np
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(42)

# Simulate 100 days of stock prices
days = np.arange(1, 101)
price = 100 + np.cumsum(np.random.randn(100) + 0.5)  # random walk with slight upward trend

# Calculate daily returns
daily_return = np.diff(price) / price[:-1]

# Key metrics
highest = price.max()
lowest = price.min()
avg_return = np.mean(daily_return)
volatility = np.std(daily_return)

# Plot stock price with markers
plt.figure(figsize=(12,6))
plt.plot(days, price, color='green', marker='o', label='Stock Price')
plt.title("Simulated Stock Price Over 100 Days")
plt.xlabel("Day")
plt.ylabel("Price ($)")
plt.legend()
plt.show()

# Plot daily returns
plt.figure(figsize=(12,5))
plt.bar(days[1:], daily_return, color='orange')
plt.title("Daily Returns")
plt.xlabel("Day")
plt.ylabel("Return")
plt.show()

# Print insights (resume-style)
print("=== Stock Simulation Insights ===")
print(f"Highest Price: ${highest:.2f}")
print(f"Lowest Price: ${lowest:.2f}")
print(f"Average Daily Return: {avg_return:.4f}")
print(f"Volatility (Std Dev of Returns): {volatility:.4f}")
print("\nThis mini-project demonstrates skills in Python, data simulation, visualization, and basic financial analysis.")
```
